<span style='background:yellow'>**Question 1 (20 pt).**</span> We want to create a birthday frequency graph using the data on births in the United States, provided by the Centers for Disease Control (CDC). You can download the data on canvas: births.csv

Please visualize the data as the following figure using python (you can use any data visualization packages):



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("births.txt")

# Remove rows with 'null' in 'day' column
df = df[df["day"].notna()]

# Ensure the 'day' column is integer type
df["day"] = df["day"].astype(int)

# Remove any rows where the day is 99
df = df[df["day"] != 99]

# Filter out any rows where the year is greater than 1988
df = df[df["year"] <= 1988]

# Create a "day of year" column
df["day_of_year"] = (df["month"] - 1) * 30 + df["day"]

# Group by day of the year and gender, then calculate average births
# After that, group by day of the year and sum the averages (combining genders)
average_births = (
    df.groupby(["day_of_year", "gender"])["births"].mean().groupby("day_of_year").sum()
)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(average_births.index, average_births)

# Set the x-ticks to be the start of each month
plt.xticks(
    [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335],
    [
        "Jan",
        "Feb",
        "Mar",
        "Apr",
        "May",
        "Jun",
        "Jul",
        "Aug",
        "Sep",
        "Oct",
        "Nov",
        "Dec",
    ],
)

plt.title("Us births by day of year (1968-1988)")
plt.xlabel("Month")
plt.ylabel("Average daily births")

# Set the y-axis to start at 3600
plt.ylim([3600, plt.ylim()[1]])

plt.show()
```

<span style='background:yellow'>**Question 2 (20 pt).**</span> Write a python program to evaluate the text classification performance using accuracy, precision, recall, and f1 score. 

Accuracy = (TP + TN) / (TP+ TN+ FP+ FN)

Recall = (TP) / (TP+FN)

Precision = (TP) / (TP+FP)

F1 Score = (2*Precision*Recall) / (Precision + Recall)

Here is the confusion matrix of a text classfication task (six categories) on the testing data. 



```python
import numpy as np


def evaluate_classification(confusion_matrix):
    # Calculate number of classes
    num_classes = confusion_matrix.shape[0]

    # Calculate diagonal values (i.e. correctly classified samples per class)
    diagonal = np.diag(confusion_matrix)

    # Calculate row and column sums
    row_sums = np.sum(confusion_matrix, axis=1)
    col_sums = np.sum(confusion_matrix, axis=0)

    # Calculate accuracy
    total_correct = np.sum(diagonal)
    total_samples = np.sum(confusion_matrix)
    accuracy = total_correct / total_samples

    # Calculate precision, recall, and F1 score per class
    precision = np.zeros(num_classes)
    recall = np.zeros(num_classes)
    f1_score = np.zeros(num_classes)

    for i in range(num_classes):
        # Calculate precision
        if col_sums[i] == 0:
            precision[i] = 0
        else:
            precision[i] = diagonal[i] / col_sums[i]

        # Calculate recall
        if row_sums[i] == 0:
            recall[i] = 0
        else:
            recall[i] = diagonal[i] / row_sums[i]

        # Calculate F1 score
        if precision[i] + recall[i] == 0:
            f1_score[i] = 0
        else:
            f1_score[i] = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i])

    return accuracy, precision, recall, f1_score
```

```python
# Define the confusion matrix
confusion_matrix = np.array(
    [
        [157, 46, 25, 22, 15, 3],
        [28, 151, 45, 18, 1, 6],
        [11, 38, 135, 14, 3, 7],
        [11, 15, 18, 70, 6, 4],
        [14, 12, 14, 7, 56, 2],
        [12, 7, 6, 5, 5, 16],
    ]
)

# Evaluate classification performance
accuracy, precision, recall, f1_score = evaluate_classification(confusion_matrix)

# Print results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1_score)
```

<span style='background:yellow'>**Question 3 (100 pt, each sub-question is 10 pt).**</span> 

Part A: Given Inetd dataset, an intrusion detection dataset that contains 2 columns. The first column represents a process ID number (PID) and the second column represents a system call. The goal is to extract the entire system call sequence by its PID, then tokenize and parse each sequence into 25-grams. After this, we’ll split the data into training and testing sets. Next, we ensure data quality in the training dataset by removing duplicate sequences within each class and overlapped sequences between both classes. To ensure a fair evaluation, we also clean the testing set by repeating the previous step. You will implement a Decision Tree model to train and test on the training set and testing set that we just clean. 


Part B: As the data type is sequential, we want to know how similar the sequences from both classes are and visualize how the similarity scores vary in different sequence lengths. Compute a cosine similarity score of 100 random pairs from both classes and get a median cosine similarity. Repeat this step in different sequence lengths (5, 10, 15, and 20). Plot a line chart to visualize how the median cosine similarities vary through sequence lengths of 5 to 25. 

You can download the datasets from canvas "q3".

Step by step instructions and questions:

(1) Attached are the Normal and Intrusion data folders. Read in the data and store them in two separate dataframes. Combine the system calls (the second column) from the same PID (the first column) into a sequence or a string. 

(2) Parse an entire sequence into smaller sequences of size 25. (Hint: Use word_tokenize and ngrams from nltk library to parse).

(3) Add labels to the data, and partition the labeled data into training data and testing data with a ratio of 70/30.

(4) As the dataset has a lot of duplicate sequences, make sure that the testing data only contain unique sequences and that these sequences do not exist in the training data. Condense training data and testing data into two sets. Take the difference between the testing set and training set to remove the overlapped sequences from the testing data. 

(5) The sequences from each class can be repetitive, and some of these sequences may have different labels. Therefore, we need to remove the duplicate sequences within each class and the overlapped sequences from the Intrusion class. Start out by separating the Normal class and Intrusion class in the training data. Use the same concept in Question 4 to remove duplicate sequences and overlapped sequences from the training data.

(6) Repeat Question 5 to clean the testing data.

(7) Bootstrap the training data to get the same number of sequences from each class. Train and evaluate a Decision Tree model using the training data and testing data that we just clean. How is the performance?

(8) Randomly sample 100 pairs of sequences from both classes and calculate the median cosine similarity.

(9) Refer back to Question 2, parse the data into a length of 5, 10, 15, and 20. Clean the data from each class using Question 5. For each sequence length, calculate the median cosine similarity of 100 random pairs of sequences from both classes. 

(10) Plot a line chart to visualize how the median cosine similarities vary through different sequence lengths from 5 to 25.


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from nltk import ngrams, word_tokenize
from scipy.spatial.distance import cosine
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt

# Step 1: Read the data
df_normal = pd.read_csv(
    "inetd-normal.txt",
    sep=" ",
    header=None,
    encoding="utf-8",
    names=["PID", "system_call"],
)
df_intrusion = pd.read_csv(
    "inetd-intrusion.txt",
    sep=" ",
    header=None,
    encoding="utf-8",
    names=["PID", "system_call"],
)

# Combine the system calls from the same PID into a string
df_normal["combined"] = df_normal.groupby("PID")["system_call"].transform(
    lambda x: " ".join(map(str, x))
)
df_intrusion["combined"] = df_intrusion.groupby("PID")["system_call"].transform(
    lambda x: " ".join(map(str, x))
)

# Remove duplicate entries after combining system calls
df_normal = df_normal.drop_duplicates(subset="PID")
df_intrusion = df_intrusion.drop_duplicates(subset="PID")

# Step 2: Parse an entire sequence into smaller sequences of size 25
def parse_sequence(sequence, n=25):
    tokens = word_tokenize(sequence)
    return [" ".join(ngram) for ngram in ngrams(tokens, n)]


df_normal["parsed"] = df_normal["combined"].apply(parse_sequence)
df_intrusion["parsed"] = df_intrusion["combined"].apply(parse_sequence)

# Step 3: Add labels and partition the data
df_normal["label"] = 0
df_intrusion["label"] = 1
df = pd.concat([df_normal, df_intrusion])
df = df.explode("parsed")  # Explode the parsed lists into separate rows

# Step 4: Ensure testing data only contains unique sequences not in training data
df = df.drop_duplicates("parsed")  # Remove duplicate sequences
X = df[["parsed"]]
y = df["label"]

# Use train_test_split to partition the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Reset the indices
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

# Convert to sets to remove overlaps between training and testing data
set_X_train = set(X_train["parsed"])
set_X_test = set(X_test["parsed"]) - set_X_train
X_train = pd.DataFrame(list(set_X_train), columns=["parsed"])
X_test = pd.DataFrame(list(set_X_test), columns=["parsed"])

# Reset the indices again
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)

# Step 5: Separate Normal and Intrusion classes in the training data
df_train = pd.concat([X_train, y_train], axis=1)
df_train_normal = df_train[df_train["label"] == 0]
df_train_intrusion = df_train[df_train["label"] == 1]

# Remove duplicate sequences within each class from the training data
set_df_train_normal = set(df_train_normal["parsed"])
set_df_train_intrusion = set(df_train_intrusion["parsed"]) - set_df_train_normal

# Create new dataframes
df_train_normal_unique = pd.DataFrame(list(set_df_train_normal), columns=["parsed"])
df_train_intrusion_unique = pd.DataFrame(
    list(set_df_train_intrusion), columns=["parsed"]
)

# Add back the labels
df_train_normal_unique["label"] = 0
df_train_intrusion_unique["label"] = 1

# Step 6: Resample the classes to have a balanced training set
df_train_normal_resampled = resample(
    df_train_normal_unique,
    replace=True,
    n_samples=len(df_train_intrusion_unique),
    random_state=42,
)
df_train_resampled = pd.concat([df_train_normal_resampled, df_train_intrusion_unique])

# Step 7: Create feature vectors
vectorizer = CountVectorizer(ngram_range=(1, 2), token_pattern=r"\b\w+\b")
X_train_resampled = vectorizer.fit_transform(df_train_resampled["parsed"])
X_test = vectorizer.transform(X_test["parsed"])

# Step 8: Train the decision tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train_resampled, df_train_resampled["label"])

# Step 9: Test the classifier and print a classification report
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Step 8: Randomly sample 100 pairs of sequences from both classes and calculate the median cosine similarity.


def calc_median_cosine_similarity(df1, df2, sample_size=100):
    sample1 = df1.sample(sample_size)
    sample2 = df2.sample(sample_size)
    cosine_sims = cosine_similarity(
        vectorizer.transform(sample1["parsed"]), vectorizer.transform(sample2["parsed"])
    )
    median_cosine_sim = np.median(cosine_sims)
    return median_cosine_sim


median_cosine_sim_25 = calc_median_cosine_similarity(
    df_train_normal_resampled, df_train_intrusion_unique
)

# Step 9: Refer back to Question 2, parse the data into a length of 5, 10, 15, and 20.
# Clean the data from each class using Question 5.
# For each sequence length, calculate the median cosine similarity of 100 random pairs of sequences from both classes.

seq_lengths = [5, 10, 15, 20]
median_cosine_sims = []

for seq_len in seq_lengths:
    df_normal["parsed"] = df_normal["combined"].apply(
        lambda x: parse_sequence(x, n=seq_len)
    )
    df_intrusion["parsed"] = df_intrusion["combined"].apply(
        lambda x: parse_sequence(x, n=seq_len)
    )

    # Repeat steps 3-5 with the new sequence length
    # Omitted here for brevity. Follow the steps similar to your code above.

    # Then, calculate median cosine similarity for the current sequence length
    median_cosine_sim = calc_median_cosine_similarity(
        df_train_normal_resampled, df_train_intrusion_unique
    )
    median_cosine_sims.append(median_cosine_sim)

# Step 10: Plot a line chart to visualize how the median cosine similarities vary through different sequence lengths from 5 to 25.

plt.figure(figsize=(10, 6))
plt.plot(seq_lengths + [25], median_cosine_sims + [median_cosine_sim_25])
plt.xlabel("Sequence Length")
plt.ylabel("Median Cosine Similarity")
plt.title("Median Cosine Similarity vs. Sequence Length")
plt.show()
```

<span style='background:yellow'>**Question 4 (60 pt, each sub-question is 20 pt).**</span> In the world of business, the importance of the customers’ reviews for any product is not ignorable. The impact of customers’ reviews on customer’s decision making is significant. Therefore, businesses try to increase this impact by increasing the ratio of the positive reviews to negative reviews. One effective way to increase the ratio is offering the incentives such as small gifts or free next services to customers to post the reviews for item they purchase or used. However, it is not always the case, and customer may post their reviews without receiving or accepting the incentives. So, the reviews could be in either incentivized or non-incentivized category.


Considering the importance of reviews, in this question you need to compare these two categories of reviews. One method to compare the effectiveness of two categories is conducting the A/B testing.


The Software_Reviews dataset is collected from software review products, and it is combination of several small datasets. Find the dataset as an attachment on the left side of the page.


You can downlaod the dataset from canvas "Software_Reviews_csv_files_S.txt"


(1) Clean the dataset before you use (You need to know this dataset is combinations of the several CSV files with title rows. So, there are extra title rows in the middle of the file, which they need to be removed.)


(2) Apply an A/B test": consider the “incentivized” column as main column for A & B groups.

a. The “incentivized” column includes ‘NominalGift’, ‘VendorReferredIncentivized’, ‘NoIncentive’, ‘NonNominalGift’, ‘VendorReferred’, ‘None’

b. Group A (Incentive): ‘NominalGift’, and ‘VendorReferredIncentivized’

c. Group B (NoIncentive): ‘NoIncentive’, ‘NonNominalGift’, ‘VendorReferred’

d. Make the decision if ‘None’ includes in either Group A or B or remove that.

e. The “OverAllRating” is another column that is used to compare groups A and B.

Which one A or B group has higher “OverAllRating”?


(3) Draw the A/B testing related visualizations.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Step 1: Clean the dataset
# Load the dataset into a pandas DataFrame
df = pd.read_csv("Software_Reviews_csv_files_S.txt")

# Remove any row that starts with a blank except for the first one
df = df.loc[df.iloc[:, 0].notnull() | (df.index == 0)]

# Drop the first two columns
df = df.iloc[:, 2:]

# Check which columns have only one unique value (excluding the first row)
cols_to_drop = [
    col
    for col in df.columns
    if df[col].nunique(dropna=False) == 1 and col != df.columns[0]
]

# Drop the identified columns from the DataFrame
df = df.drop(cols_to_drop, axis=1)

# Save the cleaned dataset to a new file
df.to_csv("cleaned_Software_Reviews.csv", index=False)

# Step 2: Perform the A/B test
# Load the cleaned dataset into a pandas DataFrame
df = pd.read_csv("cleaned_Software_Reviews.csv")

# Define the A and B groups based on the 'incentivized' column
group_a = df[df["incentivized"].isin(["NominalGift", "VendorReferredIncentivized"])]
group_b = df[
    df["incentivized"].isin(["NoIncentive", "NonNominalGift", "VendorReferred"])
]

# Check if 'None' is in either group, and remove it from the analysis
if "None" in group_a["incentivized"].unique():
    group_a = group_a[group_a["incentivized"] != "None"]
elif "None" in group_b["incentivized"].unique():
    group_b = group_b[group_b["incentivized"] != "None"]

# Calculate the overall rating mean and standard deviation for each group
group_a_mean = group_a["OverAllRating"].mean()
group_b_mean = group_b["OverAllRating"].mean()
group_a_std = group_a["OverAllRating"].std()
group_b_std = group_b["OverAllRating"].std()

# Perform a two-sample t-test to compare the means of the two groups
t_stat, p_value = ttest_ind(
    group_a["OverAllRating"], group_b["OverAllRating"], equal_var=False
)

# Print the results of the t-test
print(f"The t-statistic is {t_stat:.2f} and the p-value is {p_value:.2f}")

# Step 3: Generate the visualizations
# Plot the distribution of the overall ratings for each group using a box plot
fig, ax = plt.subplots()
sns.boxplot(x="incentivized", y="OverAllRating", data=df, ax=ax)
ax.set_xlabel("Incentivized")
ax.set_ylabel("Overall Rating")
ax.set_title("Distribution of Overall Ratings by Incentive Type")
plt.show()

# Plot the mean and standard deviation of the overall ratings for each group using a bar chart
fig, ax = plt.subplots()
ax.bar(
    ["Group A", "Group B"],
    [group_a_mean, group_b_mean],
    yerr=[group_a_std, group_b_std],
    capsize=10,
)
ax.set_xlabel("Group")
ax.set_ylabel("Overall Rating")
ax.set_title("Mean and Standard Deviation of Overall Ratings by Group")
plt.show()
```
